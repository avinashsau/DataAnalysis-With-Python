## Task 1
Using the arange() function, create an array of numbers from 0 and 10. The shape of the array must be (2, 5).

```python
import numpy as np
arr = np.arange(0,10).reshape(2,5)
print("The array created is : \n",arr)

```

## Task 2
Write a code to change the shape of the array (question 1) to (5, 2). Create a new variable for this array. Now, add this array to the original array in question 1. The resulting array should have shapes (2, 5). Save this as a variable.

```python
# to reshape this array you can use reshape() method from NumPY
arr1 = arr.reshape(5,2)
print("\n The array reshaped as : \n",arr1)

# Now if you add "arr1" with "arr" array, you will get brodcast error because they have different shapes. since we want the resulting array in shape of (2,5), we will have to transpose the "arr1".

arr2 = arr1.transpose() + arr
arr2 
```

## Task 3
Swap the indexes of the array (question 2) using the swapaxis() method. Create a new variable for the swapped array.

```python
#  The np.swapaxes() function is used to swap two axes of an
#  array, altering its layout. Unlike transpose, which can
#  rearrange the entire axes order, np.swapaxes() specifically
#  focuses on swapping two axes. To use np.swapaxes(), you
#  need to specify the axes you want to interchange.
#  For example, np.swapaxes(arr, axis1, axis2) swaps axis-1 with
#  axis-2 in the array, arr. Here is the code below:

swapped_arr = np.swapaxes(arr2, 0, 1)
swapped_arr
```

## Task 4
Write a code to slice row 1 from the swapped array in question 3.

```python
row1 = swapped_arr[0]
print(row1)

#  OR

row1 = swapped_arr[0,:]
print(row1)

#  You can see from the output that the first row has been sliced. Since we want the first row, we pass 0 (the row we
#  want is sitting at index 0) as the first number. The comma (,) means we are now going inside the row. We pass (:)
#  because we want all of the elements in a row. We combine the two to form [0, :].
```

## Task 5
You are given the following data:



Each row in the "goals" list represents the goals scored by each player in the last 5 seasons. For example, Robin’s goals are [12, 15, 16, 15, 13] and he plays for "Man UTD," and Leo's are [26, 30, 31, 25, 24] and he plays for "Barca." Using NumPy, Seaborn, and Matplotlib, create a heatmap of the goals and calculate the gradient of the data. Annotate your heat map. The x-ticks will be the names of the players, and the y-ticks will be the teams of the players.


```python
import seaborn as sns
import matplotlib.pyplot as plt

players = ["Robin", "Leo", "Pogba", "Diego", "Ronaldo"]
teams = ["Man UTD", "Barca", "Juve", "Napo", "RMadrid"]

goals = [[12,15,16,15,13],
         [26,30,31,25,24],
         [10,12,8,6,13],
         [18,19,17,20,21],
         [21,32,25,21,22]
        ]

## Creating Array from List of Goals
array1 = np.array(goals)

## COmbining Player Names with Team Label

player_team_label = [f"{team} {player}" for team , player in zip(teams, players)]
#player_team_label

fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(array1, annot=True, cmap="YlGnBu")
ax.set_xticks(np.arange(len(players)), labels=players)
ax.set_yticks(np.arange(len(teams)), rotation='horizontal', labels=player_team_label)

plt.title("Goals Scored by Players in 5 Seasons", fontsize=20)
plt.show
```
