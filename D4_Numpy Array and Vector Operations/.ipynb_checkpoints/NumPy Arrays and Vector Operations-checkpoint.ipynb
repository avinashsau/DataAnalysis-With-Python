## Task 1 - 
Create two arrays of random integers between 10 and 20. The shape of the array must be (2, 3) for the first array and (1, 3) for the second array. Check the shape of the arrays. Ensure that the results are reproducible.

```python
import numpy as np

rng = np.random.default_rng(seed=24)
arr1 = rng.integers(low=10, high=21, size=(2,3))
arr2 = rng.integers(low=10, high=21, size=(1,3))
print("Array1 = \n", arr1)
print("Array2 = \n", arr2)
print("Shape of Array1 is ", arr1.shape)
print("Shape of Array2 is ", arr2.shape)
```

## Task 2
Write a code to add the two arrays you just created in question 1. Check the shape of the resulting array. Explain why the resulting shape is that of the bigger array from the two arrays created in question 1.

```python
arr1 + arr2
```

```python
arr3 = np.add(arr1, arr2)
print("Array 3 \n", arr3)
```

## Task 4
Using NumPy, create a 1-dimensional array of 100 random integers from 0 to 10. Use the array to create a histogram of the data and calculate the median and mode. Ensure that the results are reproducible.

```python
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=42)
arr5 = rng.integers(low=0, high=11, size=(100))
arr5

plt.figure(figsize=(12,8))
plt.hist(arr5, bins=10)
plt.title("Hist Plot", fontsize=20)
plt.show
```

```python
# calculate Median & Mode
# median = np.median(arr5)  --------------- and then print median or you can directly use it in the print statement.
print("Median is : ", np.median(arr5))
```

```python
# Calculating Mode
from statistics import mode
mode = mode(arr5)

#print("Mode is :", np.mode(arr5)) -----------------------DOesnot work, giving error
print("Mode is :", mode)
```

## Task 5
Using NumPy, create a 3-dimensional array of 100 random floats between 0 and 1. Use the array to create a 3D scatter plot of the data. Ensure that your results are reproducible.

```python
import matplotlib.pyplot as plt

# Generating Data
rng = np.random.default_rng(seed=42)
#random_data = rng.random(size=(100,100,100))
#random_data
random_data = rng.integers(low=0, high=11, size=(100,3))
#arr6 = rng.random(size=(100,100,100))

# Create the FIgure and 3D Axes Object
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection='3d')

# Plot the data using the scatter() method
ax.scatter(random_data[:,0],
           random_data[:,1],
           random_data[:,2])

plt.title("3D Scatter Plot", fontsize = 20)
plt.show()
```
